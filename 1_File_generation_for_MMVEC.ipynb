```r
Male = subset_samples(alz_nt, Sex=="M")
Male = subset_samples(Male, SampleType=="Fecal")
M_Ctrl = subset_samples(Male, Treatment=="Control")
M_Abx = subset_samples(Male, Treatment=="Abx")
M_AbxFmt = subset_samples(Male, Treatment=="Abx_FMT")
```

```r
M_ALL = merge_phyloseq(M_Ctrl,M_Abx,M_AbxFmt)
M_ALL.otu = data.frame(otu_table(M_ALL))
M_ALL.meta = data.frame(sample_data(M_ALL))

write.csv(M_ALL.otu, "/home/holutz/projects/Alzheimers/mmvec_Feb2021/OTU_tables/M_ALL.otu.csv")
write.csv(M_ALL.meta, "/home/holutz/projects/Alzheimers/mmvec_Feb2021/OTU_tables/M_ALL.meta.csv")
```
